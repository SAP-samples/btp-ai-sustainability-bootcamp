#### aggregate by 8h shifts

```python
df['recStartedAt']=df['Timestamp']
df['Timestamp']=pd.to_datetime(df['Timestamp'])
df=df.fillna('-')
df['shift']=0
df['hour']=df['Timestamp'].apply(lambda x: x.hour)
df.loc[df['hour']>=8,'shift']=1
df.loc[df['hour']>=16,'shift']=2

df['date']=df['recStartedAt'].apply(lambda x: x.split(' ')[0])
df['recEndedAt']=df['Timestamp'].apply(lambda x : str(x+timedelta(hours=1)))
df
```

```python
df['MachineStatus'].unique()
```

```python
df['BROKEN']=0
df.loc[df['MachineStatus']=='BROKEN','BROKEN']=1
df['MAINTENANCE']=0
df.loc[df['PlantStatus']==0,'MAINTENANCE']=1
```

```python
df['RUNNING']=0
df.loc[df['PlantStatus']==1,'RUNNING']=1
df['FAULT']=0
df.loc[df['MachineNoise']==1,'FAULT']=1
```

```python
conditions=['BROKEN','MAINTENANCE','RUNNING','FAULT']
measures=['yield','defectiveProd','energyCons']

aggregations={}
for c in conditions:
    aggregations[c]=np.max
aggregations['recStartedAt']=np.min
aggregations['recEndedAt']=np.max

out1=df.groupby(['date','shift','plant'],as_index=False).agg(aggregations)

out2=df.drop_duplicates(['Timestamp','plant'])
out2=out2.groupby(['date','shift','plant'],as_index=False)[measures].sum()
```

```python
out1['plantStatus']='Normal'
out1.loc[   out1['BROKEN']==1 , 'plantStatus' ]= 'Break Down'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==1 ) , 'plantStatus' ]= 'Maintenance'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==0 ) & ( out1['FAULT']==1 ) & ( out1['date']>='2021' )  , 'plantStatus' ]= 'Fault'
```

```python
out1_cols=['recStartedAt', 'recEndedAt','plant', 'plantStatus','date','shift']
out2_cols=['date','shift','plant']+measures
```

```python
out1=pd.merge(out1[out1_cols] ,out2[out2_cols],how='left',on=['date','shift','plant'])
out1
```

```python
out1=out1.reset_index().rename(columns={'index':'ID'})
out1['ID']=out1['ID']+1
```

```python
out2=df.groupby(['date','shift','Machine'],as_index=False).agg(aggregations)
out2=out2[['recStartedAt', 'recEndedAt','date','Machine']+conditions].rename(columns={'Machine':'equipment_nr'})
out2.head(5)
```

```python
out2['toEquipmentStatus_code']='OK'
out2.loc[ (out2['BROKEN']==0) & (out2['MAINTENANCE']==1)  , 'toEquipmentStatus_code']='MT'
out2.loc[ (out2['FAULT']==1) & (out2['RUNNING']==1) & (out2['recStartedAt']>='2021') ,'toEquipmentStatus_code']='FL'
out2.loc[ out2['BROKEN']==1, 'toEquipmentStatus_code']='BD'
```

```python
out2.columns
```

```python
out2=pd.merge(out2[['recStartedAt', 'recEndedAt', 'date', 'equipment_nr','toEquipmentStatus_code']],out1[['recStartedAt','ID']],how='left',on='recStartedAt')
out2=out2.rename(columns={'ID':'plantCond_ID'})
```

```python
out2=out2.reset_index().rename(columns={'index':'ID'})
```

```python
out2
```

```python
###V1
import glob
import random

sound_list=glob.glob('../ai-models/predictive-maintenance/LGPsound/anomaly*/*.wav')

detectedAt=[]
detectedDate=[]
equipment=[]
rawValue=[]

for idx,row in anomalies.iterrows():
    for t in range(0,6):
        
        clip='/'.join(random.choice(sound_list).split('/')[-2:])
        rawValue.append('/media/LGPsound/{}'.format(clip))       
        detectedAt.append(row['Timestamp']+timedelta(minutes=10*t)) 
        detectedDate.append(row['detectedDate']) 
        equipment.append(row['Machine'])

        
        
ID=[i for i in range(1,len(rawValue)+1)]
anomalyType_code=['A1'if c.split('/')[-2]=='anomaly1'  else 'A2' for c in rawValue ]
confidence=[np.random.uniform(0.7,0.98) for i in rawValue]   
status=[1 for i  in rawValue]
```

```python
###V2
import glob
import random
import datetime

image_list1=glob.glob('../ai-models/defect-detection/data/Images/NG/*')
image_list2=glob.glob('../ai-models/defect-detection/data/Images/OK/*')

image=[]
qualityLabel=[]
detectedAt=[]
detectedDate=[]
perc=[]

for idx,row in defects.iterrows():
    if int(row['defectiveProd']) >0 :
        T=row['Timestamp']
        for M in range(0,6):

            R = np.random.uniform()
            if R > 0.1:
                im='/'.join(random.choice(image_list2).split('/')[-2:])
                q='Y'
            else:
                im='/'.join(random.choice(image_list1).split('/')[-2:])
                q='N'
            image.append('/media/cv/{}'.format(im))
            detectedAt.append(T) 
            qualityLabel.append(q)
            detectedDate.append(row['detectedDate']) 
            T=T+timedelta(minutes=10)
        
ID=[i for i in range(1,len(image)+1)]
plant=[1010 for i in range(1,len(image)+1) ]
plantSection=['YOH' for i in range(1,len(image)+1)]
productId=['SG23' for i in range(1,len(image)+1)]
productName=['LGP' for i in range(1,len(image)+1)]
confidence=[np.random.uniform(0.8,0.98)for i in range(1,len(image)+1) ]   
```
