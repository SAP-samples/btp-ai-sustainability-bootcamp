#### aggregate by 8h shifts

```python
df['recStartedAt']=df['Timestamp']
df['Timestamp']=pd.to_datetime(df['Timestamp'])
df=df.fillna('-')
df['shift']=0
df['hour']=df['Timestamp'].apply(lambda x: x.hour)
df.loc[df['hour']>=8,'shift']=1
df.loc[df['hour']>=16,'shift']=2

df['date']=df['recStartedAt'].apply(lambda x: x.split(' ')[0])
df['recEndedAt']=df['Timestamp'].apply(lambda x : str(x+timedelta(hours=1)))
df
```

```python
df['MachineStatus'].unique()
```

```python
df['BROKEN']=0
df.loc[df['MachineStatus']=='BROKEN','BROKEN']=1
df['MAINTENANCE']=0
df.loc[df['PlantStatus']==0,'MAINTENANCE']=1
```

```python
df['RUNNING']=0
df.loc[df['PlantStatus']==1,'RUNNING']=1
df['FAULT']=0
df.loc[df['MachineNoise']==1,'FAULT']=1
```

```python
conditions=['BROKEN','MAINTENANCE','RUNNING','FAULT']
measures=['yield','defectiveProd','energyCons']

aggregations={}
for c in conditions:
    aggregations[c]=np.max
aggregations['recStartedAt']=np.min
aggregations['recEndedAt']=np.max

out1=df.groupby(['date','shift','plant'],as_index=False).agg(aggregations)

out2=df.drop_duplicates(['Timestamp','plant'])
out2=out2.groupby(['date','shift','plant'],as_index=False)[measures].sum()
```

```python
out1['plantStatus']='Normal'
out1.loc[   out1['BROKEN']==1 , 'plantStatus' ]= 'Break Down'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==1 ) , 'plantStatus' ]= 'Maintenance'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==0 ) & ( out1['FAULT']==1 ) & ( out1['date']>='2021' )  , 'plantStatus' ]= 'Fault'
```

```python
out1_cols=['recStartedAt', 'recEndedAt','plant', 'plantStatus','date','shift']
out2_cols=['date','shift','plant']+measures
```

```python
out1=pd.merge(out1[out1_cols] ,out2[out2_cols],how='left',on=['date','shift','plant'])
out1
```

```python
out1=out1.reset_index().rename(columns={'index':'ID'})
out1['ID']=out1['ID']+1
```

```python
out2=df.groupby(['date','shift','Machine'],as_index=False).agg(aggregations)
out2=out2[['recStartedAt', 'recEndedAt','Machine']+conditions].rename(columns={'Machine':'equipment_nr'})
out2.head(5)
```

```python
out2['equipmentStatus']='OK'
out2.loc[ (out2['BROKEN']==0) & (out2['MAINTENANCE']==1)  , 'equipmentStatus']='MN'
out2.loc[ (out2['FAULT']==1) & (out2['RUNNING']==1) & (out2['recStartedAt']>='2021') ,'equipmentStatus']='FL'
out2.loc[ out2['BROKEN']==1, 'equipmentStatus']='BD'
```

```python
out2.columns
```

```python
out2=pd.merge(out2[['recStartedAt', 'recEndedAt', 'equipment_nr','equipmentStatus']],out1[['recStartedAt','ID']],how='left',on='recStartedAt')
out2
```

```python
out2=out2.reset_index().rename(columns={'index':'ID'})
```

```python
out2.to_csv(out2_path,index=False,sep=';')
```

```python
out1.to_csv(out1_path,index=False,sep=';')
```

```python
defects=df[['Timestamp','defectiveProd']].drop_duplicates(['Timestamp'])
```

```python
defects=defects[defects['Timestamp'].as_type(str)>='2021']
```

```python
defects['detectedDate']=defects['Timestamp'].apply(lambda x: str(x).split(' ')[0])
defects=defects[defects['detectedDate']>='2021']
defects
```

```python
import glob
import random

image_list=glob.glob('../ai-models/defect-detection/data/Images/NG/*')
image=[]
detectedAt=[]
detectedDate=[]
for idx,row in defects.iterrows():
    for d in range(0,int(row['defectiveProd'])):
        im=random.choice(image_list).split('/')[-1]
        image.append('/media/cv/{}'.format(im))
        detectedAt.append(row['Timestamp']) 
        detectedDate.append(row['detectedDate']) 

        
        
ID=[i for i in range(1,len(image)+1)]
plant=[1010 for i in range(1,len(image)+1) ]
plantSection=['YOH' for i in range(1,len(image)+1)]
productId=['SG23' for i in range(1,len(image)+1)]
productName=['LGP' for i in range(1,len(image)+1)]
qualityLabel=['N' for i in range(1,len(image)+1)]
confidence=[np.random.uniform(0.8,0.98)for i in range(1,len(image)+1) ]   
```

```python
out3=pd.DataFrame(data={'ID':ID \
                       ,'detectedAt':detectedAt \
                       ,'detectedDate':detectedDate \
                       ,'plant':plant \
                       ,'plantSection':plantSection \
                       ,'productId':productId \
                       ,'productName':productName \
                       ,'qualityLabel':qualityLabel \
                       ,'confidence':confidence \
                       ,'image':image \
                       })
```

```python
out3.to_csv(out3_path,index=False)
```

```python
CO2= 0.0301  # kg/kW
```

```python
#compute additional columns
df['effectiveYield']= df['yield']-df['defectiveProd']
df['PlantCO2Emissions']= df['energyCons']*CO2

df['Timestamp']=pd.to_datetime(df['Timestamp'])
df=df.fillna('-')
df['year_month']=df['Timestamp'].apply(lambda x: str(x.year)+'-'+str(x.month).zfill(2))

plant_cols=[ c for c in cols if 'plant' in c ]
#machine_cols=[ c for c in cols if 'Machine' in c ]
plant_measures=['yield','defectiveProd','effectiveYield','energyCons','PlantCO2Emissions']
#machine_measures=['MachineEnergyConsumption (kW)','MachineCO2Emissions', 'PlantEnergyCost']

#aggregate monthly
dfm=df.drop_duplicates(subset=['Timestamp']+plant_cols)
dfm=dfm.groupby(['year_month','plant'],as_index=False)[plant_measures].sum()
```
