```python
df['recStartedAt']=df['Timestamp']
df['Timestamp']=pd.to_datetime(df['Timestamp'])
df=df.fillna('-')
df['shift']=0
df['hour']=df['Timestamp'].apply(lambda x: x.hour)
df.loc[df['hour']>=8,'shift']=1
df.loc[df['hour']>=16,'shift']=2

df['date']=df['recStartedAt'].apply(lambda x: x.split(' ')[0])
df['recEndedAt']=df['Timestamp'].apply(lambda x : str(x+timedelta(hours=1)))
df
```

```python
df['BROKEN']=0
df.loc[df['MachineStatus']=='BROKEN','BROKEN']=1
df['MAINTENANCE']=0
df.loc[df['PlantStatus']==0,'MAINTENANCE']=1
```

```python
df['RUNNING']=0
df.loc[df['PlantStatus']==1,'RUNNING']=1
df['FAULT']=0
df.loc[df['MachineNoise']==1,'FAULT']=1
```

```python
conditions=['BROKEN','MAINTENANCE','RUNNING','FAULT']
measures=['yield','defectiveProd','energyCons']

aggregations={}
for c in conditions:
    aggregations[c]=np.max
aggregations['recStartedAt']=np.min
aggregations['recEndedAt']=np.max

out1=df.groupby(['date','shift','plant'],as_index=False).agg(aggregations)

out2=df.drop_duplicates(['Timestamp','plant'])
out2=out2.groupby(['date','shift','plant'],as_index=False)[measures].sum()
```

```python
out1['plantStatus']='Normal'
out1.loc[   out1['BROKEN']==1 , 'plantStatus' ]= 'Break Down'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==1 ) , 'plantStatus' ]= 'Maintenance'
out1.loc[ ( out1['BROKEN']==0 ) & ( out1['MAINTENANCE']==0 ) & ( out1['FAULT']==1 ) & ( out1['date']>='2021' )  , 'plantStatus' ]= 'Fault'
```

```python
out1_cols=['recStartedAt', 'recEndedAt','plant', 'plantStatus','date','shift']
out2_cols=['date','shift','plant']+measures
```

```python
out1=pd.merge(out1[out1_cols] ,out2[out2_cols],how='left',on=['date','shift','plant'])
out1
```

```python
out1=out1.reset_index().rename(columns={'index':'ID'})
out1['ID']=out1['ID']+1
```

```python
out1.to_csv(out1_path,index=False,sep=';')
```
